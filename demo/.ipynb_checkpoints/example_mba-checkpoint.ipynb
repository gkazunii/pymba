```python
import sys
sys.path.append("../src/")
import numpy as np
from mproject import MBA
```

```python
# Prepare random input non-negative normalized tensor
import utils_test
N = 8
P = utils_test.generate_low_rank_tensor( (N,N,N,N), 10, seed=None)
P = P / np.sum(P)
D = np.ndim(P)
```

```python
# Let us run two-body Tsallis deformed decomposition with q=0.5 with Newton method
body = 3
Q, theta, eta, his = MBA(P, body, lr_search=True, Newton=True, max_iter=100, epsilon_auto=True,
                         chi="Tsallis", q=0.5, 
                         verbose_interval=1, verbose=True);
```

```python
# Let us run three-body Kaniadakis deformed decomposition with k=0.5 with Newton method
body = 3
Q, theta, eta, his = MBA(P, body, lr_search=True, Newton=True, max_iter=100, epsilon_auto=True,
                         chi="Kani", k=0.5, 
                         verbose_interval=1, verbose=True);
```

```python
# define interaction with all one-body interactions and
# two-body interactions of (1,2) and (1,4) and
# three-body interactions of (1,2,3).
intracts = [ [1,1,1,1],[1,0,1,0,0,0],[1,0,0,0],[0] ]
Q, theta, eta, his = MBA(P, intracts, lr_search=True, Newton=True, max_iter=100, epsilon_auto=True,
                         chi="Tsallis", q=0.5, 
                         verbose_interval=1, verbose=True);
```
